```python
import tkinter as tk
from tkinter import filedialog
import moviepy.editor as mp

# Create a Tkinter window to show the file dialog box
root = tk.Tk()
root.withdraw()

# Bring the Tkinter window to the front to avoid issues with pop-up blockers
root.call('wm', 'attributes', '.', '-topmost', True)

# Show the file dialog box and allow the user to select a video file
file_path = filedialog.askopenfilename(filetypes=[("Video Files", "*.mp4;*.avi;*.mkv;*.wmv;*.flv")])

if file_path:
    # Create a VideoFileClip object from the selected video file
    video = mp.VideoFileClip(file_path)

    # Extract audio from the video file
    audio = video.audio

    # Show another file dialog box to allow the user to select the output audio file
    save_path = filedialog.asksaveasfilename(defaultextension='.mp3')

    if save_path:
        # Write the audio to the output file
        audio.write_audiofile(save_path)
```
